```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DailyDelhiClimateTrain.csv')
```

```python
df = df[['date','meantemp']]
```

```python
df = df.iloc[1:101,:]
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.5).mean()
```

```python
df['EWMA'] = x1
```

```python
plt.scatter(df['date'],df['meantemp'],color='red')
plt.plot(df['date'],x1,color='black')
plt.show()
```
