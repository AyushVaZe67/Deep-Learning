```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Function to generate a spiral dataset
def generate_spiral_data(samples_per_class=50, num_classes=2, noise=0.2):
    np.random.seed(42)
    X = []
    y = []
    for class_num in range(num_classes):
        for i in range(samples_per_class):
            r = i / samples_per_class  # radius
            theta = class_num * np.pi + r * 2 * np.pi  # angle
            x1 = r * np.sin(theta) + np.random.randn() * noise
            x2 = r * np.cos(theta) + np.random.randn() * noise
            X.append([x1, x2])
            y.append(class_num)
    
    X = np.array(X)
    y = np.array(y)
    return X, y

# Generate the dataset
X, y = generate_spiral_data(samples_per_class=50, num_classes=2)

# Convert to DataFrame
df = pd.DataFrame(X, columns=['X', 'Y'])
df['class'] = y

# Save to CSV (optional)
df.to_csv("spiral_dataset.csv", index=False)

# Plot the dataset
plt.scatter(df['X'], df['Y'], c=df['class'], cmap='viridis')
plt.show()
```
