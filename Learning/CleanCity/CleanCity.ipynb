```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load your trained YOLOv8 model (use the path to your best.pt)
model = YOLO('best.pt')
```

```python
img_path = r"E:\AYUSH VAZE\DEGREE\Study\6th sem\Others\Sakshi BE Project\allwaste1.png"
img = cv2.imread('bottle1.jpg')
results = model(img_path)[0]
```

```python
# cv2.imshow('window',img)
# cv2.waitKey(0)
```

```python
import cv2
import matplotlib.pyplot as plt
from collections import Counter

# Copy original image
img = results.orig_img.copy()

# Extract all class IDs and labels
class_ids = [int(box.cls[0]) for box in results.boxes]
labels = [results.names[class_id] for class_id in class_ids]

# Count each label
label_counts = Counter(labels)
total_count = sum(label_counts.values())

# === Print Summary to Console ===
print("Detected Items Summary:")
for label, count in label_counts.items():
    print(f'- {label}: {count}')
print(f'\nTotal Items Detected: {total_count}')

# Threshold logic
threshold = 10
status_message = "Kachara gaadi ko bulao" if total_count > threshold else "No need"
print(f'Status: {status_message}')  # Show message in console

# === Draw Results on Image ===
for box in results.boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    class_id = int(box.cls[0])
    label = results.names[class_id]
    conf = float(box.conf[0])
    text = f'{label} {conf:.2f}'

    # Draw bounding box and label
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 1)
    cv2.putText(img, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                0.5, (0, 255, 0), 1)

# Draw total count
cv2.putText(img, f'Total Count: {total_count}', (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

# Draw status message
status_color = (0, 0, 255) if total_count > threshold else (0, 200, 0)
cv2.putText(img, status_message, (10, 70),
            cv2.FONT_HERSHEY_SIMPLEX, 0.6, status_color, 2)

# Convert image and display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.title('YOLOv8 Detection Result', fontsize=12)
plt.show()
```
