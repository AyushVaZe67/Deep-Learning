```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load your trained YOLOv8 model (use the path to your best.pt)
model = YOLO('best.pt')
```

```python
img_path = r"E:\AYUSH VAZE\DEGREE\Study\6th sem\Others\Sakshi BE Project\vege.jpg"
img = cv2.imread('vege.jpg')
```

```python
cv2.imshow('window',img)
cv2.waitKey(0)
```

```python
results = model(img_path)[0]
```

```python
import cv2
import matplotlib.pyplot as plt

# Assuming 'results' is from YOLOv8 inference
img = results.orig_img.copy()

for box in results.boxes:
    x1, y1, x2, y2 = map(int, box.xyxy[0])
    label = results.names[int(box.cls[0])]
    conf = float(box.conf[0])
    text = f'{label} {conf:.2f}'

    # Draw rectangle
    cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Draw text with smaller font scale
    cv2.putText(img, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                fontScale=0.5, color=(0, 255, 0), thickness=1)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(12, 8))
plt.imshow(img_rgb)
plt.axis('off')
plt.title('YOLOv8 Detection Result', fontsize=12)
plt.show()
```
