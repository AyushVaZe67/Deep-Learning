```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

%matplotlib inline

# Load your trained YOLOv8 model (use the path to your best.pt)
model = YOLO('runs/detect/train8/weights/best.pt')

# Load an image
img_path = r'C:\Users\devan\deadlift-4\test\images\go3_mp4-89_jpg.rf.855bb3573f86ad1a614feca874bb897e.jpg'
img = cv2.imread(img_path)

# Run detection (inference)
results = model(img_path)[0]  # get first batch result

# Print detected boxes, classes, confidences
print("Boxes (x1, y1, x2, y2):")
print(results.boxes.xyxy.cpu().numpy())

print("Class IDs:")
print(results.boxes.cls.cpu().numpy())

print("Confidences:")
print(results.boxes.conf.cpu().numpy())

# To display the image with bounding boxes using OpenCV (or matplotlib)
# YOLOv8 has built-in plotting function:
img_with_boxes = results.plot()  # returns an image with boxes drawn (numpy array)

# Convert BGR to RGB for matplotlib
img_with_boxes = cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB)

# Display
plt.figure(figsize=(12,8))
plt.imshow(img_with_boxes)
plt.axis('off')
plt.title('YOLOv8 Detection Result')
plt.show()
```

```python
import cv2
from ultralytics import YOLO
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time

model = YOLO('runs/detect/train8/weights/best.pt')
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

max_frames = 500  # max frames to run
frame_count = 0

try:
    while frame_count < max_frames:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        results = model(frame)[0]

        # results.plot() returns frame with boxes drawn (highlighted)
        frame_with_boxes = results.plot()

        # Convert BGR to RGB for matplotlib display
        frame_rgb = cv2.cvtColor(frame_with_boxes, cv2.COLOR_BGR2RGB)

        # Clear output and display the image
        clear_output(wait=True)
        plt.imshow(frame_rgb)
        plt.axis('off')
        plt.title(f"Frame {frame_count + 1}")
        plt.show()

        frame_count += 1
        time.sleep(0.05)  # adjust for smoother/slower display

except KeyboardInterrupt:
    print("Stopped by user")

cap.release()
```
