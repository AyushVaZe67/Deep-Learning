```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load your trained YOLOv8 model (use the path to your best.pt)
model = YOLO('best.pt')
```

```python
img_path = r"E:\AYUSH VAZE\DEGREE\Study\Data Science\DL\Projects\LockOut\deadlift detection\AyushTry\deadpic (8).png"
img = cv2.imread('deadpic (8).png')

results = model(img_path)[0]  # get first batch result

# Print detected boxes, classes, confidences
print("Boxes (x1, y1, x2, y2):")
print(results.boxes.xyxy.cpu().numpy())

print("Class IDs:")
print(results.boxes.cls.cpu().numpy())

print("Confidences:")
print(results.boxes.conf.cpu().numpy())


img_with_boxes = results.plot()  # returns an image with boxes drawn (numpy array)

img_with_boxes = cv2.cvtColor(img_with_boxes, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(12,8))
plt.imshow(img_with_boxes)
plt.axis('off')
plt.title('YOLOv8 Detection Result')
plt.show()
```
